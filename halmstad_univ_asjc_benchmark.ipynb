```python
#install/import libraries
import pandas as pd
!pip install pandasql
from pandasql import sqldf
import plotly.graph_objects as go
import plotly.express as px
import math
```

```python
# Load scopus data
asjc_all = pd.read_excel('asjc_sweden_scopus.xlsx')
```

```python
#group data by year
by_year = asjc_all.groupby(asjc_all.year)['eid'].nunique().reset_index()
by_year.plot(kind='bar', x='year', legend=None, title='Publications per year', figsize=(10, 10))

#save to excel
by_year.to_excel('publications_per_year.xlsx', index=False)
```

```python
#group data by field
by_field = asjc_all.groupby(asjc_all.description)['eid'].count().reset_index().sort_values('eid')
by_field.plot(kind='barh', x='description', legend=None, title='Publications per asjc field', figsize=(15, 10))
#save to excel
by_field.to_excel('publications_per_field.xlsx', index=False)
```

```python
# Yearly counts by asjc field
yearly_counts = asjc_all.groupby(['year', 'description'])['eid'].count().reset_index()
yearly_pivot = yearly_counts.pivot(index='description', columns='year', values='eid')
yearly_pivot = yearly_pivot.fillna(0)
yearly_pivot

#save to excel
yearly_pivot.to_excel('yearly_pivot.xlsx', index=False)
```

```python
# limit to data from Hamlmstad university (hh) ID in Scopus: 60022085 (for id:s to other organisations, see file: org_id_scopus.xlsx on GitHub)
query = """SELECT*
           FROM asjc_all 
           WHERE affiliations LIKE '%60022085%'
        """
# Execute the query
asjc_publ_hh = sqldf(query)
asjc_publ_hh

#save the data to excel
asjc_publ_hh.to_excel('asjc_publications_hh.xlsx')
```

```python
#group data by year (hh)
by_year_hh = asjc_publ_hh.groupby(asjc_publ_hh.year)['eid'].nunique().reset_index()
by_year_hh.plot(kind='bar', x='year', legend=None, title='Publications per year', figsize=(10, 10))
#save to excel
by_year_hh.to_excel('publications_per_year_hh.xlsx', index=False)
```

```python
#group data by field
by_field_hh = asjc_publ_hh.groupby(asjc_publ_hh.description)['eid'].count().reset_index().sort_values('eid')
by_field_hh.plot(kind='barh', x='description', legend=None, title='Publications per asjc field', figsize=(15, 10))
#save to excel
by_field_hh.to_excel('publications_per_field_hh.xlsx', index=False)
```

```python
# Yearly counts by asjc field (hh)
yearly_counts = asjc_publ_hh.groupby(['year', 'description'])['eid'].count().reset_index()
yearly_pivot = yearly_counts.pivot(index='description', columns='year', values='eid')
yearly_pivot = yearly_pivot.fillna(0)
yearly_pivot

#save to excel
yearly_pivot_hh.to_excel('yearly_pivot_hh.xlsx', index=False)
```
